```python

messages = [
    {"role": "system", "content":"You are a helpful assistant named Jarvis."}
]

def chat_with_ollama(userPrompt:str, model:str = "steamdj/llama3.1-cpu-only:latest", should_stream=False):
    messages.append({"role": "user", "content": userPrompt})
    model_response = ollama.chat(model=model, messages=messages, stream=should_stream)

    if should_stream:
        return model_response  # handled in stream_chat_with_ollama
    else:
        messages.append({"role": "assistant", "content": model_response.message.content})
        return model_response.message.content



def stream_chat_with_ollama(user_prompt: str) -> None:
    try:
        stream = chat_with_ollama(user_prompt, should_stream=True)
        print("Generating Answer.....\n\n")
        assistant_response = []

        for chunk in stream:
            if chunk.message and chunk.message.content:
                assistant_response.append(chunk.message.content)
                print(chunk.message.content, end="", flush=True)

        full_response = "".join(assistant_response)
        if full_response.strip():
            messages.append({"role": "assistant", "content": full_response})
            print("\n\nAssistant message: ", full_response, "\n")

    except Exception as e:
        print(f"\n\nAn error occurred: {e}\n")



if __name__ == "__main__":
    while(True):
        print("You are now chatting with ollama.\n\n")
        user_prompt = input("Youüë¶: ")
        if user_prompt == "q":
            break
        print("\nBotü§ñ: \n")
        stream_chat_with_ollama(user_prompt=user_prompt)
    
    for c in messages:
        print(c)
```

```python
def generate_with_ollama(prompt:str, model:str = "steamdj/llama3.1-cpu-only:latest")->None:
    try:
        res = ollama.generate(model=model, prompt=prompt, stream=False, options={"temperature" : 0})
        print(res.response)
    except Exception as e:
        print("Some Error occured",e,"\n")
        raise e

def raw_chat_with_ollama():
    print("You are now chatting with ollama. This chat is stateless\n\n")
    while(True):
        try:
            user_prompt = input("Youüë¶: ")
            if user_prompt == "q" or user_prompt == "Q":
                break
            else:
                generate_with_ollama(user_prompt)
        except Exception as e:
            print(f"Some Error Occured: \n {e}\n")
            break
raw_chat_with_ollama()
```
